```python
#5.What is the trend of confirmed cases over time globally?
corona_confirm=corona.groupby(by='ObservationDate')['Confirmed'].sum().sort_values()
corona_confirm.plot(kind='line',title='trend of confirmed cases over time globally',figsize=(15,6),ylabel='confirmed cases')
```

```python
#6.Which province/state has reported the highest number of confirmed cases?
corona.groupby(by='Province/State')['Confirmed'].max().sort_values(ascending=False).head(1)
```

```python
#7.Which country/region has the highest number of deaths?
corona.groupby(by='Country/Region')['Deaths'].max().sort_values(ascending=False).head(1)
```

```python
#8.How does the number of confirmed cases vary across different provinces/states?
confirm_cases=corona.groupby(by='Province/State')['Confirmed'].sum().sort_values(ascending=False).head(10)
confirm_cases.plot(kind='bar',title=' confirmed cases vary across different provinces/states',ylabel='conformed cases')
```

```python
#9.What is the trend of deaths over time globally?
death_trend=corona.groupby('ObservationDate')['Deaths'].sum()
death_trend.plot(title=' trend of deaths over time globally',color='pink')
```

```python
#10.Which country/region has the highest number of recovered cases?
corona.groupby('Country/Region')['Recovered'].sum().sort_values(ascending=False).head(1)
```

```python
#11.How does the number of recovered cases vary across different countries/regions?
recover_cases=corona.groupby('Country/Region')['Recovered'].sum().sort_values(ascending=False).head(10)
recover_cases.plot(kind='bar')
```

```python
#15.How does the mortality rate vary across different countries/regions?
mortality=corona.groupby('Country/Region')[['Confirmed','Deaths']].sum()
mortality['Mortality_rate']=(mortality['Deaths']/mortality['Confirmed'])*100    
mortality.sort_values(by='Mortality_rate',ascending=False).head(10)
```

```python
#16.How does the recovery rate vary across different countries/regions?
corona_recove=corona.groupby(by='Country/Region')[['Recovered','Confirmed']].sum()
corona_recove['Recovery_rate']=(corona_recove['Recovered'] / corona_recove['Confirmed']) * 100
corona_recove.sort_values(by='Recovery_rate',ascending=False).head(20)
```

```python
#18.How does the fatality rate vary across different provinces/states?
fatality=corona.groupby('Province/State')[['Deaths','Confirmed']].sum()
fatality['fatality_rate']=(fatality['Deaths'] / fatality['Confirmed']) * 100
fatality.sort_values(by='fatality_rate',ascending=False).head(20)
```

```python
#19.How does the recovery rate vary across different provinces/states?
recover=corona.groupby('Province/State')[['Confirmed','Recovered']].sum()
recover['Recovery Rate']=(recover['Recovered'] / recover['Confirmed']) * 100
recover.sort_values(by='Recovery Rate',ascending=False).head(20)
```
